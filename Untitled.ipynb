```python
import patch
patch.all()

import pytorch_lightning as pl
from pytorch_lightning.callbacks import (
    ModelCheckpoint,
)

import torchvision
import torch

import os
import yaml
import sys

from config import Config
from data import get_data, CustomDataset, CustomDataModule
from transformations import get_pipeline, Pipeline, Replicate
from models import ResnetEncoder
from utils import get_subset_of_size, TimerPrintCallback,  HParamsSaveCallback

import self_supervised

import matplotlib.pyplot as plt 

torch.backends.cuda.matmul.allow_tf32 = True
torch.backends.cudnn.allow_tf32 = True

def main(config: Config):
    identities = config.data.classes

    default_root_dir = os.path.join(
            config.data.output,
            'tensorboard',
            f'data_{config.data.name}',
            f'model_{config.model.name}',
            f'transform_{config.transformations.type}',
            config.name,
    )

    train_pipeline, eval_pipeline, collate_fn = get_pipeline(config.transformations)

    train_loaders, all_val_loaders, all_test_loaders = get_data(
        config, 
        train_pipeline, 
        eval_pipeline, 
        collate_fn,
    )
    
    val_loader = all_val_loaders[0][0]

    for X, _ in val_loader:
        X = torch.stack(X)
        _, axs = plt.subplots(X.shape[1], X.shape[0], figsize=(7,4))

        for i,ax in zip(X.transpose(0,1).flatten(0,1),axs.ravel()):
            ax.imshow(i.transpose(0,1).transpose(1,2), vmin=i.min(), vmax=i.max())
        break
    return val_loader

if __name__  ==  '__main__':
    config = 'config.yaml'
    if config == '-':
        config = yaml.load(sys.stdin, Loader = yaml.SafeLoader)
    else:
        config = yaml.load(open(config, 'r'), Loader = yaml.SafeLoader)

    configs_to_run = [Config(**c) for c in config['runs']]
    if len(sys.argv) >  1:
        try:
            index = int(sys.argv[1])
            configs_to_run = [configs_to_run[index]]
        except:
            print('failed to parse argv')

    for run in configs_to_run:
        val_loader = main(run)
```

```python
import patch
patch.all()

import torch
from transformations import LogPolar

import matplotlib.pyplot as plt 

def main():
    data = [
        # torch.abs(torch.arange(10000).reshape(1,100,100)),
        torch.abs(5000-torch.arange(10000).reshape(1,100,100)),
        # torch.abs(torch.arange(10000).reshape(1,100,100).transpose(1,2)),
        torch.abs(5000-torch.arange(10000).reshape(1,100,100).transpose(1,2)),
    ]
    _, axs = plt.subplots(ncols=4, nrows=len(data), figsize=(8,5))
    for idx, img in enumerate(data):
        lp_in = LogPolar(input_shape = (img.shape[-2:]), output_shape=(190,165), smoothing=0, mask=False, position='inscribed')
        lp_out = LogPolar(input_shape = (img.shape[-2:]), output_shape=(190,165), smoothing=None, mask=False, position='circumscribed')
        lp_out_mask = LogPolar(input_shape = (img.shape[-2:]), output_shape=(190,165), smoothing=0, mask=True, position='circumscribed')

        i = axs[idx][0].imshow(img.transpose(0,1).transpose(1,2), cmap='gray', vmin = img.min(), vmax = img.max())
        axs[idx][1].imshow(lp_in(img).transpose(0,1).transpose(1,2), cmap='gray', vmin = img.min(), vmax = img.max())
        axs[idx][2].imshow(lp_out(img).transpose(0,1).transpose(1,2), cmap='gray', vmin = img.min(), vmax = img.max())
        axs[idx][3].imshow(lp_out_mask(img).transpose(0,1).transpose(1,2), cmap='gray', vmin = img.min(), vmax = img.max())
        
        for ax in axs[idx]:
            ax.set_xticks([])
            ax.set_yticks([])
        
        axs[idx][0].set_title('Original', fontsize=10)
        axs[idx][1].set_title('LP inscribed', fontsize=10)
        axs[idx][2].set_title('LP circumscribed',fontsize=10)
        axs[idx][3].set_title('LP circumscribed, masked',fontsize=10)
    # plt.colorbar(i)
if __name__  ==  '__main__':
    main()
```
